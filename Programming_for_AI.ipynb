# **we will create and plot sin() cos() in 2 different subplots**

```python
import numpy as np
import matplotlib.pyplot as plt

xstart = 0
xstop = 2*np.pi
increment = 0.1

x = np.arange(xstart , xstop , increment)

y = np.sin(x)
z = np.cos(x)

print(y)
print(z)

plt.subplot(2,1,1)
plt.plot(x,y)
plt.xlabel("x",color='r',fontsize=18)
plt.ylabel("sin(x)",color='r',fontsize=18)
plt.title("sin(angle)",color='r',fontsize=18)
plt.grid()
plt.show()

plt.subplot(2,1,2)
plt.plot(x,z)
plt.xlabel("x",color='r',fontsize=18)
plt.ylabel("cos(x)",color='r',fontsize=18)
plt.title("cos(angle)",color='r',fontsize=18)
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

xstart = 0
xstop = 2*np.pi
increment = 0.1

x = np.arange(xstart , xstop , increment)
y = np.cosh(x)
z = np.tan(x)

print(y)
print(z)

plt.subplot(2,1,1)
plt.plot(x,y,color='g')
plt.xlabel("x",color='g',fontsize=18)
plt.ylabel("cosh(x)",color='g',fontsize=18)
plt.title("cos hyperbolic fn",fontsize=24)
plt.grid()
plt.show()

plt.subplot(2,1,2)
plt.plot(x,z,color='r')
plt.xlabel("x",color='g',fontsize=18)
plt.ylabel("tan(x)",color='g',fontsize=18)
plt.title("tan(Angle)",fontsize=24)
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

xstart = 0
xstop = 2*np.pi
increment = 0.1

x = np.arange(xstart , xstop , increment)
y = np.sin(x)
z = np.cos(x)

print(y)
print(z)

plt.subplot(2,1,1)
plt.plot(x,y,color='g')
plt.plot(x,z,color='r')
plt.xlabel("x",color='g',fontsize=18)
plt.ylabel("sin(x) cos(x)",color='g',fontsize=18)
plt.title("sin(fn) cos(fn)",fontsize=24)
plt.grid()
plt.show()

plt.subplot(2,1,2)
plt.plot(x,z,color='r')
plt.xlabel("x",color='g',fontsize=18)
plt.ylabel("cos(x)",color='g',fontsize=18)
plt.title("cos(fn)",fontsize=24)
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

xstart = 0
xstop = 2*np.pi
increment = 0.1

x = np.arange(xstart,xstop,increment)
y = np.sin(x)
z1 = np.cos(x)
z2 = np.tanh(x)
z3 = np.sinh(x)

print(y)
print(z1)
print(z2)

plt.subplot
plt.plot(x,y,color='b')
plt.plot(x,z1,color='r')
plt.plot(x,z2,color='g')
plt.xlabel("x",color='b',fontsize=18)
plt.ylabel("sin(fn) cos(fn) tanh(fn)",color='b',fontsize=18)
plt.title("TRIGONOMETRIC & TRIGONOMETRIC HYPERNOLIC FUNCTION",color='b',fontsize=24)
plt.grid()
plt.show()

plt.subplot
plt.plot(x,y,color='b')
plt.plot(x,z3,color='g')
plt.xlabel("x",color='b',fontsize=18)
plt.ylabel("sin(fn) sinh(fn)",color='b',fontsize=18)
plt.title("TRIGONOMETRIC & TRIGONOMETRIC HYPERNOLIC FUNCTION", color='b',fontsize=24)
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

xstart = 0
xstop = 2*np.pi
increment = 0.1

x = np.arange(xstart,xstop,increment)
y = np.sin(x)
z = np.sin(-x)
zx1 = np.cos(x)
zx2 = np.cos(-x)

print(y)
print(z)

plt.subplot(2,1,1)
plt.plot(x,y,color='m')
plt.plot(x,z,color='b')
plt.grid()
plt.show()

plt.subplot(2,1,2)
plt.plot(x,zx1,color='g')
plt.plot(x,zx2,color='y')
plt.grid()
plt.show()

plt.subplot(2,1,2)
plt.plot(x,zx2,color='r')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

xstart = 0
xstop = 2*np.pi
increment = 0.1

x = np.arange(xstart,xstop,increment).reshape(-1,1)
y = np.sinh(x)
z1 = np.cos(x)
z2 = np.tanh(x)

print(y)
print(z1)
print(z2)

plt.subplot(2,1,1)
plt.plot(x,y,color='m')
plt.plot(x,z1,color='b')
plt.grid()
plt.show()

plt.subplot(2,1,2)
plt.plot(x,z1,color='m')
plt.plot(x,z2,color='b')
plt.grid()
plt.show()
```
